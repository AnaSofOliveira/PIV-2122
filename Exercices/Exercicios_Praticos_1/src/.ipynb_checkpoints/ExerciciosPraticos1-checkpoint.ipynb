### Exercício 1.2 - Leitura e visualização de um vídeo ou imagens de uma câmara 

```python
video = cv2.VideoCapture(0)
fps = video.get(cv2.CAP_PROP_FPS)

frames = 0
while(True):
      
    flag, frame = video.read()
    
    if(flag):
        cv2.imshow("frame", frame)
        #frames =+ 1
    
    key = cv2.waitKey(int(1000/fps))

    if(key == 27): 
        cv2.destroyAllWindows()
        video.release()
        break
    
```

### Exercício 2.1 – Cromakey (blue screening)
#### ImageOut = objectImage x Mask + background x !Mask

```python
falcao = cv2.imread(src + "falcon.jpg")
mascara = cv2.imread(src + "mask.png")
fundo = cv2.imread(src + "florest.jpg")
mascara_invertida = 255-mascara

corpo_falcao = cv2.multiply(falcao, mascara, scale=1/255)
floresta = cv2.multiply(fundo, mascara_invertida, scale=1/255)

resultado = cv2.add(corpo_falcao, floresta)

cv2.imshow("Falcao", falcao) 
cv2.imshow("Mascara", mascara) 
cv2.imshow("Fundo", fundo) 
cv2.imshow("Mascara Invertida", mascara_invertida) 

cv2.imshow("Corpo Falcao", corpo_falcao) 
cv2.imshow("Floresta Sem Falcao", floresta) 

cv2.imshow("Resultado Final", resultado) 

cv2.imwrite(src + "resultados/resultado.jpg", resultado)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Exercício 3.1 - Histograma de uma imagem

```python
falcon = cv2.imread(src + "falcon.jpg")

channels = [0, 1, 2]

for channel in channels: 
    # Imagem dentro de lista, canal em lista [0 - B, 1 - G, 2 - R], máscara, [número de pontos do histograma], [range do histograma]
    histogram = cv2.calcHist([falcon], [channel], None, [256], [0, 255])

    # Histograma vem como uma matriz. Podemos transformar em array unidimensional
    hist = np.squeeze(histogram)

    plt.bar(np.arange(256), hist) # adicionar color
    plt.show()
```

### Exercício 3.3 - Operadores morfológicos

#### Elementos Estruturantes (Kernels)

```python
rect_kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (5,5))
eliptical_kernal = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5,5))
crossed_kernal = cv2.getStructuringElement(cv2.MORPH_CROSS, (5,5))
```

#### Erosão

```python
fingerprint = cv2.imread(src + "fingerprint.jpg", 0)
brain = cv2.imread(src + "BrainMRI_Axial.jpg", 0)

thresh_finger, fingerprint = cv2.threshold(fingerprint, 127, 255, cv2.THRESH_OTSU)
thresh_brain, brain = cv2.threshold(brain, 127, 255, cv2.THRESH_OTSU)

fingerprint_rect = cv2.erode(fingerprint, rect_kernel)
fingerprint_elip = cv2.erode(fingerprint, eliptical_kernal)
fingerprint_cross = cv2.erode(fingerprint, crossed_kernal)

brain_rect = cv2.erode(brain, rect_kernel)
brain_elip = cv2.erode(brain, eliptical_kernal)
brain_cross = cv2.erode(brain, crossed_kernal)

cv2.imshow("Fingerprint", fingerprint)
cv2.imshow("Fingerprint With Rect Kernel", fingerprint_rect)
cv2.imshow("Fingerprint With Eliptical Kernel", fingerprint_elip)
cv2.imshow("Fingerprint With Crossed Kernel", fingerprint_cross)

cv2.imshow("Brain", brain)
cv2.imshow("Brain With Rect Kernel", brain_rect)
cv2.imshow("Brain With Eliptical Kernel", brain_elip)
cv2.imshow("Brain With Crossed Kernel", brain_cross)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

#### Dilatação

```python
fingerprint = cv2.imread(src + "fingerprint.jpg", 0)
brain = cv2.imread(src + "BrainMRI_Axial.jpg", 0)

thresh_finger, fingerprint = cv2.threshold(fingerprint, 127, 255, cv2.THRESH_OTSU)
thresh_brain, brain = cv2.threshold(brain, 127, 255, cv2.THRESH_OTSU)

fingerprint_rect = cv2.dilate(fingerprint, rect_kernel)
fingerprint_elip = cv2.dilate(fingerprint, eliptical_kernal)
fingerprint_cross = cv2.dilate(fingerprint, crossed_kernal)

brain_rect = cv2.dilate(brain, rect_kernel)
brain_elip = cv2.dilate(brain, eliptical_kernal)
brain_cross = cv2.dilate(brain, crossed_kernal)

cv2.imshow("Fingerprint", fingerprint)
cv2.imshow("Fingerprint With Rect Kernel", fingerprint_rect)
cv2.imshow("Fingerprint With Eliptical Kernel", fingerprint_elip)
cv2.imshow("Fingerprint With Crossed Kernel", fingerprint_cross)

cv2.imshow("Brain", brain)
cv2.imshow("Brain With Rect Kernel", brain_rect)
cv2.imshow("Brain With Eliptical Kernel", brain_elip)
cv2.imshow("Brain With Crossed Kernel", brain_cross)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Exercício 3.4 - Labeling

```python
imagem = cv2.imread(src + "ImTest1.jpg")
imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

cv2.imshow("Imagem Original", imagem)
cv2.imshow("Imagem Cinzenta", imagem_cinza)


# Find Contours
ret, thresh = cv2.threshold(imagem_cinza, 127, 255, cv2.THRESH_OTSU)
contornos, hierarquia = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)
cv2.drawContours(imagem, contornos, -1, [0,255,0], 3)

cv2.imshow("Imagem Com Contornos", imagem)


# Componentes Conexos

numRegioes, labels = cv2.connectedComponents(thresh)

#print("Numero de Regiões: ", numRegioes)
#print("Labels: ", labels)

colorMap = psColor.CreateColorMap(numRegioes, 1)
pseudoC = psColor.Gray2PseudoColor(labels, colorMap)

cv2.imshow("Imagem Color Mapped", pseudoC)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Exercício 3.5 - Extracção de características

```python
coin_names = ["P1000697s.jpg", "P1000698s.jpg", "P1000699s.jpg", "P1000703s.jpg", 
             "P1000705s.jpg", "P1000706s.jpg", "P1000709s.jpg", "P1000710s.jpg", 
             "P1000713s.jpg"]

def mostrarImagem(eixo, titulo, imagem):
    eixo.axis("off")
    eixo.title.set_text(titulo)
    eixo.imshow(cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB))
    
for name in coin_names: 
    moedas_path = src + "conjuntoTreino/" + name

    fig, ax = plt.subplots(4, 2, figsize=(20,20))
    plt.suptitle("Imagem " + name)
    
    # Leitura da imagem
    moedas_original = cv2.imread(moedas_path)
    #cv2.imshow("Moedas Original", moedas_original)
    mostrarImagem(ax[0,0], "Moedas Original", moedas_original)
    
    # Conversão para tons de cinzento
    moedas_cinza = cv2.split(moedas_original)[2]
    #cv2.imshow("Moedas Cinza", moedas_cinza)
    mostrarImagem(ax[0,1], "Moedas Cinza", moedas_cinza)

    # Binarização
    thresh, moedas_binarizadas = cv2.threshold(moedas_cinza, np.min(moedas_cinza), np.max(moedas_cinza), cv2.THRESH_OTSU)
    #cv2.imshow("Moedas Binarizadas", moedas_binarizadas)
    mostrarImagem(ax[1,0], "Moedas Binarizadas", moedas_binarizadas)

    # Melhoramento
    elementoEstruturante = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (6,6))
    moedas_fechadas = cv2.morphologyEx(moedas_binarizadas, cv2.MORPH_CLOSE, elementoEstruturante)
    #cv2.imshow("Moedas Fechadas", moedas_fechadas)
    mostrarImagem(ax[1,1], "Moedas Fechadas", moedas_fechadas)
    
    elementoEstruturante = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (40,40))
    moedas_erodidas = cv2.erode(moedas_fechadas, elementoEstruturante)
    #cv2.imshow("Moedas Erodidas", moedas_erodidas)
    mostrarImagem(ax[2,0], "Moedas Erodidas", moedas_erodidas)

    elementoEstruturante = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (21,21))
    moedas_abertas = cv2.morphologyEx(moedas_erodidas, cv2.MORPH_OPEN, elementoEstruturante)
    #cv2.imshow("Moedas Abertas", moedas_abertas)
    mostrarImagem(ax[2,1], "Moedas Abertas", moedas_abertas)
    
    elementoEstruturante = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (25,25))
    moedas_dilatadas = cv2.dilate(moedas_abertas, elementoEstruturante)
    #cv2.imshow("Moedas Ditatadas", moedas_dilatadas)
    mostrarImagem(ax[3,0], "Moedas Ditatadas", moedas_dilatadas)

    

    

    #cv2.waitKey(0)
    #cv2.destroyAllWindows()
plt.show()
```

#### Exercício 6.2 - Extração de características de textura
##### 6.2.1 - Calcule a densidade de contornos;
##### NOTAS: Agarrar na imagem, passar para tons de cinzento, calcular derivada x e y e calcular angulo (arctang2 com intensidades em x e y) e intensidade do gradiente (rais do quadrado de x + quadrado de y) e calcular densidade de contornos (threshold) 


```python
imagem = cv2.imread(src + "lily-lotus-flowers.jpg")

imagem_gray = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY) 

derivada_x = cv2.convertScaleAbs(cv2.Sobel(imagem_gray, cv2.CV_16S, dx=1, dy=0, ksize=3))
derivada_y = cv2.convertScaleAbs(cv2.Sobel(imagem_gray, cv2.CV_16S, dx=0, dy=1, ksize=3))

magnitude_gradiente = np.sqrt(cv2.add(np.float32(derivada_x)**2, np.float32(derivada_y)**2))
magnitude_normalizada = magnitude_gradiente/np.max(magnitude_gradiente)

angulo = np.arctan2(derivada_y, derivada_x)
angulo_normalizado = angulo/(2*np.pi) + 0.5 #Normalização do angulo do gradiente. Como temos valores entre pi e 2*pi

cv2.imshow("Imagem Original", imagem) 
cv2.imshow("Imagem Cinzenta", imagem_gray) 
cv2.imshow("Contornos Y", derivada_x) 
cv2.imshow("Contornos X", derivada_y)

thresholds = [50, 100, 150, 200, 250]

for threshold in thresholds: 
    contornos = magnitude_gradiente.astype(np.uint8) > threshold
    cv2.imshow("Densidade th=" + str(threshold), contornos.astype(np.uint8)*255)
    
    edgeness = round(np.sum(contornos)/contornos.size, 5)
    print("Densidade th=", threshold, ": ", edgeness)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
pessoa = cv2.imread(src + "pessoa.jpg")
pessoa_YCbCr = cv2.cvtColor(pessoa, cv2.COLOR_BGR2YCR_CB)

minCrCb = np.array([0, 133, 77])
maxCrCb = np.array([255, 173, 127])
mascara = cv2.inRange(pessoa_YCbCr, minCrCb, maxCrCb)
print(mascara)

kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (5,5))
mascara_melhorada = cv2.morphologyEx(mascara, cv2.MORPH_CLOSE, kernel)

pessoa_contornada = pessoa.copy()
cv2.imshow("Pessoa Contornada1", pessoa_contornada)
contornos, hierarquia = cv2.findContours(mascara_melhorada, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
cv2.drawContours(pessoa_contornada, contornos, -1, [0,255,0], 1)


cv2.imshow("Pessoa", pessoa)
cv2.imshow("Mascara", mascara)
cv2.imshow("Mascara Melhorada", mascara_melhorada)
cv2.imshow("Pessoa Contornada", pessoa_contornada)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
